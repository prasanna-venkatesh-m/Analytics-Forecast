```python
import pandas as pd

df = pd.read_csv("/content/Cleaned_data.csv", index_col=0)

# 1. Ensure Date is in datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# 2. Sort values to ensure chronological order within each branch
df = df.sort_values(by=['BranchID', 'Date'])

# 3. Define the columns you want to lag
# Excluding BranchID and Date from being lagged themselves
features_to_lag = [
    'CountofGroup', 'CountofClientName', 'CountofAccountID',
    'CountofCreditOfficerID', 'SumofPrincipalOutstanding',
    'SumofInterestOutstanding', 'SumofTotalPrincipalOverdue',
    'SumofTotalInterestOverdue', 'SumofTotalPAR'
]

# 4. Create the lags
lags = [1, 7, 15, 30, 365]

for lag in lags:
    for col in features_to_lag:
        lag_col_name = f"{col}_{lag}day_lag"

        # Group by BranchID so lags are specific to each branch
        df[lag_col_name] = df.groupby('BranchID')[col].shift(lag)

        # 5. Requirement: If lag data is unavailable, keep the current day's data
        df[lag_col_name] = df[lag_col_name].fillna(df[col])

print(df.head())
```

```python
df.to_csv('lag_data.csv')
```
