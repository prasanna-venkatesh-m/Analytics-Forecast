```python
import pandas as pd

# Load CSV
df = pd.read_csv("/content/LatestData.csv")
df.head()
```

```python
df[df.duplicated(['BranchID', 'Date'], keep=False)].sort_values(['BranchID','Date'])
```

```python

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

# Sort
df = df.sort_values(['BranchID', 'Date'])

# Create full date range
full_dates = pd.date_range(df['Date'].min(), df['Date'].max(), freq='D')

# Create multi-index (BranchID × all dates)
branches = df['BranchID'].unique()

full_index = pd.MultiIndex.from_product(
    [branches, full_dates],
    names=['BranchID', 'Date']
)

# Reindex to include missing dates
df = df.set_index(['BranchID', 'Date']).reindex(full_index)

# Fill missing values using previous day (per branch)
df = df.groupby(level=0).ffill()

df = df.fillna(0)

# Optional: reset index
df = df.reset_index()

print(df)
```

```python
df.to_csv('output.csv')
```
